# Phase 0.4 - Dataset Folder Structure (Diabetic Retinopathy Project)

```python
# Phase 0.4 - Dataset Folder Structure (Diabetic Retinopathy Project)
import os

# Base path inside your Drive
base_dir = '/content/drive/MyDrive/datasets/diabetic_retinopathy'

# Folder names (binary classification: normal vs infected)
folders = [
    'train/normal',
    'train/infected',
    'val/normal',
    'val/infected',
    'test/normal',
    'test/infected'
]

# Create folders if not exist
for f in folders:
    path = os.path.join(base_dir, f)
    os.makedirs(path, exist_ok=True)

print("✅ Folder structure created successfully at:", base_dir)
```

# Verification of Folders

```python
for folder in ['train', 'val', 'test']:
    path = os.path.join(base_dir, folder)
    print(folder, "exists?", os.path.exists(path))
    if os.path.exists(path):
        print("  Subfolders:", os.listdir(path))
```
